```python
import numpy as np
import pandas as pd
import statistics as st
from numpy import loadtxt
from pandas import read_csv
import statistics as st
import seaborn as sns
import missingno as mnso
from numpy import nan
from numpy import isnan
from numpy import loadtxt
from pandas import read_csv
from matplotlib import pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from pandas.plotting import scatter_matrix
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
from sklearn.metrics import precision_recall_curve,plot_precision_recall_curve,plot_roc_curve

data = pd.read_csv('data.CSE708.csv')
# print(data.head())
# print(data.describe())
print(data.info())
print(data.shape)


# ## Replacing zeros with NAN
Mydata = data.replace(0, nan)
# Mydata.dropna(inplace = True)
# Mydata.fillna(Mydata.mean(), inplace=True)
#checking data set before removing missing values and outliers
mnso.bar(Mydata)
plt.show()
# #Checking number of missing values as zeros
print(Mydata.isnull().sum())


##One hot coding of Outcome
xcolumns = Mydata.drop(columns = ['Outcome'])
# print(xcolumns)
dummy_Variables = pd.get_dummies(Mydata['Outcome'])
# print(dummy_Variables.head(7))
NewData2 = pd.concat((Mydata,dummy_Variables),axis=1)
# NewData2.head(7)
NewData3 = NewData2.drop(['Outcome'], axis = 1) 
NewData4 = NewData3.drop(['R'], axis = 1)
Dataframe_Final = NewData4.rename(columns = {"N":"Outcome"})
# print(Dataframe_Final.head())
# # print(Dataframe_Final.shape)

## Outlier detection and removal
mean = Dataframe_Final.mean()
std = Dataframe_Final.std()
outrows = []
for j in range(0, len(Dataframe_Final.columns)-1):
    outlier = 0
    for i in range(0, len(Dataframe_Final.index)):
        z_score = (Dataframe_Final.iloc[i,j]-mean[j])/std[j]
        if abs(z_score) > 3:
            outlier += 1
            outrows.append(i)
outrows=set(outrows)
NewData = Dataframe_Final.drop(index=outrows, axis=0)

##Replacing remaining zeros with mean

NewData.dropna(inplace = True)
NewData.fillna(Mydata.mean(), inplace=True)
print(NewData.isnull().sum())
print(NewData.shape)

mnso.bar(NewData)
plt.show()

# class distribution
print(NewData.groupby('Outcome').size())

dataframe = NewData.copy()
dataframe.dropna(axis = 0, how = 'any', thresh = None, subset = None, inplace = True)
dataframe = dataframe.sample(frac = 1, random_state = 1)

# # Separate class
class_0 = dataframe[dataframe['Outcome'] == 0].sample(n = 150, random_state = 1)
class_1 = dataframe[dataframe['Outcome'] == 1].sample(n = 150, random_state = 1) # print the shape of the class

print('class 0:', class_0.shape)
print('class 1:', class_1.shape)

NewData = pd.concat([class_0, class_1], axis=0).sample(frac = 1, random_state = 1)



# #Normalization of data
# # for column in NewData.columns:
# #     NewData[column] = NewData[column] / NewData[column].max()


    
## Correlation matrix method for feature reduction
matrix = NewData.corr().round(3)
plt.figure(figsize=(20,20))
sns.heatmap(matrix,annot = True)
plt.show()

Corr_sum  = matrix.sum(axis=1)

Corr_sum = Corr_sum[0:30]
Corr_sum = Corr_sum.sort_values()
print(Corr_sum)

##Best attributes
Best_Attributes = Corr_sum[14:30].index
print(Best_Attributes)

Mydatabest_Att = NewData[Best_Attributes]
scatter_matrix(Mydatabest_Att,figsize  = [15, 15], color = 'magenta')
plt.show()

## print(Mydatabest_Att.head())
# print(Mydatabest_Att.shape)

## Moving forward we can use the Mydatabest_Att
## SVM Model

#create an scaler instance
scaler = StandardScaler()

array7 = Mydatabest_Att.values
X7 = array7[:,0:15]
Y7 = NewData['Outcome']
X7_train, X7_validation, Y7_train, Y7_validation = train_test_split(X7, Y7, test_size=0.2, random_state=1)

#fit and transfor the scaler
scaled_X_train = scaler.fit_transform(X7_train)
scaled_X_test = scaler.transform(X7_validation)

model = SVC(gamma='auto')
print('\nThe accuracy of the SVM model')
kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
cv_results = cross_val_score(model, scaled_X_train, Y7_train, cv=kfold, scoring='accuracy')
print('%f (%f)' % (cv_results.mean(), cv_results.std()))


# # #Split the data
X7_train, X7_validation, y7_train, y7_validation = train_test_split(X7, Y7, test_size = 0.2, random_state=1)
model = SVC()
model.fit(X7_train,y7_train)

print('*******************************************************************')
y_pred1 = model.predict(X7_validation)
print(y_pred1)

##Model Evaluation
print('Support vector machine')
confusion=confusion_matrix(Y7_validation, y_pred1)
report = classification_report(Y7_validation, y_pred1)
print(confusion)
print(f"Classification Report:\n{report}")

# # ## ROC curve
# # #plot the confusion matrix
# # plot_confusion_matrix(model,X7_validation,y7_validation)

# # # #Scaled so highest value=1
# # # plot_confusion_matrix(model,X7_validation,y7_validation)

# # # print(classification_report(y7_validation,y_pred))

plot_precision_recall_curve(model,X7_validation,y7_validation)
plot_roc_curve(model,X7_validation,y7_validation)

## Logistic Regression
print('Logistic Regression')

log_model = LogisticRegression()
log_model.fit(scaled_X_train,Y7_train)
print(log_model.coef_)
y_pred2 = log_model.predict(scaled_X_test)
print(y_pred2)
print(accuracy_score(Y7_validation,y_pred2))
print(confusion_matrix(Y7_validation,y_pred2))
print(classification_report(Y7_validation,y_pred2))
# plot_confusion_matrix(log_model,scaled_X_test,Y7_validation)
# plot_precision_recall_curve(log_model,scaled_X_test,Y7_validation)
# plot_roc_curve(log_model,scaled_X_test,Y7_validation)


## KNN Model
print('K Nearest Neighbors')
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(scaled_X_train,y7_train)
y_pred3 = knn_model.predict(scaled_X_test)
print(accuracy_score(Y7_validation,y_pred3))
print(confusion_matrix(Y7_validation,y_pred3))
report = classification_report(Y7_validation,y_pred3)
print(report)

#create an empty list to hold the errors
test_error_rates = []
for k in range(1,30):
    #classifier with new k value
    knn_model = KNeighborsClassifier(n_neighbors=k)
    #fit the KNN model
    knn_model.fit(scaled_X_train,y7_train) 
    #predict the classes of the test dataset
    y_pred_test = knn_model.predict(scaled_X_test)
    #calculate the error 
    test_error = 1 - accuracy_score(Y7_validation,y_pred_test)
    #append the error to the list
    test_error_rates.append(test_error)

#draw the line plot to identify the k value with the lowest error (elbow)
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")

#Using pipeline to perform multiple steps
print('K = 15 Nearest Neighbors')
scaler = StandardScaler()
knn5 = KNeighborsClassifier(n_neighbors=15)
#provide the steps as tuple
operations = [('scaler',scaler),('knn15',knn15)]
#The purpose of the pipeline is to assemble several steps 
pipe = Pipeline(operations)
pipe.fit(X7_train,y7_train)
pipe_pred = pipe.predict(X7_validation)
print(classification_report(Y7_validation,pipe_pred))
```
